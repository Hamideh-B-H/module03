```python
import pandas as pd
from typing import List

def analyze_main_movie_dataset(file_path: str) -> None:
    """
    Analyzes the complete Rotten Tomatoes dataset to identify data quality issues.
    Focuses on missing values for SQL constraints and messy strings for normalization.
    """
    try:
        # Load the complete file containing 17,712 films
        df = pd.read_csv(file_path)

        print(f"--- Analysis of {file_path} ---")
        print(f"Total Rows: {len(df)}")

        # 1. Identify columns with empty cells (NULLs)
        # This determines which SQL columns can be 'NOT NULL'
        print("\n--- Columns with Missing Values ---")
        missing_counts = df.isnull().sum()
        print(missing_counts[missing_counts > 0])

        # 2. Analyze Production Companies for inconsistencies
        # Helps identify why we normalized this into a separate table
        print("\n--- Distinct Production Companies (Top 30) ---")
        # Sorting helps spot "Warner Bros" vs "Warner Bros."
        companies: List[str] = sorted(df['production_company'].dropna().unique())
        for company in companies[:30]:
            print(f"- {company}")

        # 3. Inspect the Date Formats
        # We need to know if we must convert "Friday, 8 October 1937" to "1937-10-08"
        print("\n--- Sample Release Date Formats ---")
        print("Original Release Date samples:")
        print(df['original_release_date'].dropna().head(5))
        print("\nStreaming Release Date samples:")
        print(df['streaming_release_date'].dropna().head(5))

    except FileNotFoundError:
        print(f"Error: The file '{file_path}' was not found in the directory.")

if __name__ == "__main__":
    analyze_main_movie_dataset('rotten_tomatoes_movies.csv')
```

### Notes on "create_database" step
**Fulfills Normalization** : I have eliminated multivalued attributes (genres, actors, etc.) and redundant descriptive data (companies, ratings).
**Logical Structure**: My tables are organized into "Entities" and "Junctions," making the database easy to manage.
**Appropriate Data Types**: I used DATE for release dates, INTEGER for runtimes/counts, and SERIAL for auto-incrementing IDs.
**Constraints**: I applied NOT NULL only where data is guaranteed (like titles and links) and allowed NULL for the "messy" columns like critics' consensus checked above.

### Notes on "db_load" step
To load a normalized database, we cannot just insert rows; we must follow a specific order:

    -Load Entity Tables first (production_companies, content_ratings, persons, genres) to generate the IDs we need for Foreign Keys.
    -Load the movies table using those IDs.
    -Load the movie_metrics and Junction Tables using the newly created movie_id.
